## Import Library

```python
import numpy as np
import pandas as pd
import pickle
import tensorflow as tf
```

## Load Model

```python
with open('pipeline.pkl','rb') as file1:
    pipeline = pickle.load(file1)

func_imp = tf.keras.models.load_model('model functional improve.h5')
```

## Generate random data

```python
random = np.random.default_rng(seed=2712)
region_category = random.choice(['Village', 'City', 'Town'],size=20)
membership_category = random.choice(['No Membership',
                                     'Basic Membership', 
                                     'Platinum Membership', 
                                     'Gold Membership', 
                                     'Premium Membership', 
                                     'Silver Membership'],size=20)
joined_through_referral = random.choice(['Yes','No'],size=20)
preferred_offer_types = random.choice(['Credit/Debit Card Offers',
                                       'Gift Vouchers/Coupons',
                                       'Without Offers'],size=20)
medium_of_operation = random.choice(['Desktop', 'Smartphone', 'Both'],size=20)
days_since_last_login = random.integers(0,26,size=20)
avg_time_spent = random.integers(0.0,3235.5785210942604,size=20)
avg_transaction_value = random.integers(800.46,99914.05,size=20)
avg_frequency_login_days = random.integers(0.0,73.06199459430009,size=20)
points_in_wallet = random.integers(0.0,2069.069760814851,size=20)
used_special_discount = random.choice(['Yes','No'],size=20)
offer_application_preference = random.choice(['Yes','No'],size=20)
feedback = random.choice(['Too many ads', 'No reason specified', 'Reasonable Price',
                          'Quality Customer Care', 'Poor Website', 'Poor Customer Service',
                          'Poor Product Quality', 'User Friendly Website', 'Products always in Stock'],size=20)

data_inf = pd.DataFrame({'region_category' : region_category,
                        'membership_category' : membership_category,
                        'joined_through_referral' : joined_through_referral,
                        'preferred_offer_types' : preferred_offer_types,
                        'medium_of_operation' : medium_of_operation,
                        'days_since_last_login' : days_since_last_login,
                        'avg_time_spent' : avg_time_spent,
                        'avg_transaction_value' : avg_transaction_value,
                        'avg_frequency_login_days' : avg_frequency_login_days,
                        'points_in_wallet' : points_in_wallet,
                        'used_special_discount' : used_special_discount,
                        'offer_application_preference' : offer_application_preference,
                        'feedback' : feedback                       
                        })
data_inf
```

## Predict

```python
inf_final = pipeline.transform(data_inf)
inf_predict = np.round(func_imp.predict(inf_final))
inf_predict
```

```python
data_inf['Predict'] = inf_predict
data_inf
```
